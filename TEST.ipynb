```python
import datetime as dt
import numpy as np
import pandas as pd

from datetime import datetime, timedelta
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func

from flask import Flask, jsonify
```

```python

#########################################
##    Database SETUP
#########################################
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
                     

# reflect an existing database into a new model
# reflect the tables
# save reference to tables
Base = automap_base()
Base.prepare(engine, reflect=True)
Measurement = Base.classes.measurement
Station = Base.classes.station
session = Session(engine)
```

```python
lastdate = session.query(Measurement.date).order_by(Measurement.date.desc()).first()
lastdate
```

```python
last_date = datetime.strptime(lastdate[0],'%Y-%m-%d')
start_date=last_date - timedelta(days=365)
```

```python
start_date = '2016-08-23'
end_date = '2017-08-23'
    
result = session.query(Measurement.date, Measurement.prcp).\
filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()
#Create dictionary from the row data and append to a list
one_year_prcp = []
for d,p in result:
    prcp_dic={}
    prcp_dic["Date"]=d
    prcp_dic["PRCP value"]=p
    one_year_prcp.append(prcp_dic)
one_year_prcp
#     return jsonify(one_year_prcp)
```

```python
most_active = "USC00519281"
session.query(Measurement.date,Measurement.tobs).filter(Measurement.station == most_active).filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()
#Create dictionary from the row data and append to a list
one_year_tobs = []
for d,t in result:
    tob_dic={}
    tob_dic["Date"]=d
    tob_dic["Tobs Value"]=t
    one_year_tobs.append(tob_dic)
one_year_tobs
#     return jsonify(one_year_tobs)
```

```python

#################################################
# Flask Setup
#################################################
app = Flask(__name__)


#################################################
# Flask Routes
#################################################

@app.route("/")
def welcome():
    return (
        f'Available routes:<br/>'
        f'/api/v1.0/precipitation<br/>'
        f'/api/v1.0/stations<br/>'
        f'/api/v1.0/tobs<br/>'
        f'/api/v1.0/<start><br/>'
        f'/api/v1.0/<start><end><br>'
    )
                     
@app.route("/api/v1.0/stations")
def station():
#  Create our session (link) from Python to the DB
    session = Session(engine)
    station_info = session.query(Station.station, Station.name).all()
    session.close()
                     
    #covert the list of tuples into a normal list
    all_station=[]
    for x,y in station_info:
        st_dic={}
       	st_dic["station"]=x
        st_dic["name"]=y            
        all_station.append(st_dic)
    return jsonify(all_station)         

if __name__=='__main__':
    app.run(debug=True)
```
